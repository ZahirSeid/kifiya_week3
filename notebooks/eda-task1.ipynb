```python
# Load the sorted Excel file
file_path = '../Data/MachineLearningRating_v3_sorted.xlsx' 
df = pd.read_excel(file_path)
# Display basic information
df.info()

# Display the first few rows
df.head()
```

```python
# Summary statistics
df.describe()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Impute missing values for numerical columns with the mean
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].mean())

# Impute missing values for categorical columns:
# 1. For categorical variables (Bank, AccountType, etc.), fill with 'Unknown' or mode (most frequent value)
categorical_columns = df.select_dtypes(include=['object']).columns

for col in categorical_columns:
    if df[col].isnull().sum() > 0:  # Only fill if there are missing values
        mode_value = df[col].mode()[0]  # Get the most frequent value
        df[col] = df[col].fillna(mode_value)

# Impute missing data in `NewVehicle`, `WrittenOff`, `Rebuilt`, and `Converted` using the mode (for categorical nature)
df['NewVehicle'] = df['NewVehicle'].fillna(df['NewVehicle'].mode()[0])
df['WrittenOff'] = df['WrittenOff'].fillna(df['WrittenOff'].mode()[0])
df['Rebuilt'] = df['Rebuilt'].fillna(df['Rebuilt'].mode()[0])
df['Converted'] = df['Converted'].fillna(df['Converted'].mode()[0])

# For `CrossBorder` and `NumberOfVehiclesInFleet` which have all missing data:
# Since these columns are entirely missing, they can be dropped if they don't provide value for analysis
df = df.drop(columns=['CrossBorder', 'NumberOfVehiclesInFleet'])

# Verify imputation
print("Missing values after imputation:")
print(df.isnull().sum())

df.to_excel("../Data/MachineLearningRating_v3_cleaned.xlsx", index=False)
```

```python
# Distribution of Numerical Features
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Calculate the number of rows and columns needed for subplots
num_columns = len(numerical_columns)
num_rows = (num_columns // 3) + (1 if num_columns % 3 != 0 else 0)  # To handle any remainder

plt.figure(figsize=(14, 5 * num_rows))  # Adjust figure size based on the number of rows
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(num_rows, 3, i)  # Dynamically adjust the grid size
    sns.histplot(df[col], kde=True, bins=30, color='blue')
    plt.title(f'Distribution of {col}')
    plt.tight_layout()

plt.show()
```

```python
# Distribution of Categorical Features
categorical_columns = df.select_dtypes(include=['object', 'category']).columns

# Calculate the number of rows and columns needed for subplots
num_columns = len(categorical_columns)
num_rows = (num_columns // 3) + (1 if num_columns % 3 != 0 else 0)  # To handle any remainder

plt.figure(figsize=(14, 5 * num_rows))  # Adjust figure size based on the number of rows
for i, col in enumerate(categorical_columns, 1):
    plt.subplot(num_rows, 3, i)  # Dynamically adjust the grid size
    sns.countplot(x=df[col], hue=df[col], palette='viridis', legend=False)
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45)  # Rotate X-axis labels for better readability
    plt.tight_layout()

plt.subplots_adjust(hspace=0.5, wspace=0.5)  # Adjust spacing between plots
plt.show()
```
